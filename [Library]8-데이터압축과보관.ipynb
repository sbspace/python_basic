```python
# 데이터의 압축과 해제 

import zlib

data = " Bitcoin is too cheep, keep buying." * 10000
print(len(data))

compress_data = zlib.compress(data.encode(encoding='utf-8')) # 유니코드 문자열을 UTF-8 형식으로 인코딩하여 바이트로 만들고 compres 해야함.
print(len(compress_data), compress_data)

org_data = zlib.decompress(compress_data).decode('utf-8')
print(len(org_data))
```

```python
# 데이터를 압축하여 파일로 저장. 혹은 해제 

import gzip 

with gzip.open('data.txt.gz', 'wb') as f:
    f.write(data.encode('utf-8'))

with gzip.open('data.txt.gz', 'rb') as f:
    read_data = f.read().decode('utf-8')

assert data == read_data
```
